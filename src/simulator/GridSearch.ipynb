```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('training-data/AVGBD_data_r.csv', header=None, names=['Bounded Slowdown'])

plt.figure(figsize=(15, 9))
df
plt.plot(df)
```

```python
# Nombre total de lignes dans le DataFrame
total_lignes = len(df)

# Nombre de lignes dans chaque sous-DataFrame
taille_section = total_lignes // 10

# Liste pour stocker les 10 sous-DataFrames
sous_dataframes = []
plt.figure(figsize=(15, 9))

# Dcouper le DataFrame en 10 sous-DataFrames
for i in range(10):
    debut_index = i * taille_section
    fin_index = (i + 1) * taille_section
    if i == 9:
        fin_index = total_lignes
    sous_df = df.iloc[debut_index:fin_index].reset_index(drop=True)  # Rinitialiser l'index
    sous_dataframes.append(sous_df)

# Affichage des 10 sous-DataFrames
for i, sous_df in enumerate(sous_dataframes):
    plt.plot(sous_df, label=f"Tuple {i+1}")

# Ajout de la lgende
plt.xscale('linear')  # chelle linaire pour l'axe des x
plt.yscale('linear')  # chelle linaire pour l'axe des y
plt.title('Random shuffle evaluation')
plt.xlabel('Generation')
plt.ylabel('Bounded Slowdown')
plt.legend(loc='upper right', fontsize='small')

# Affichage du graphique
plt.show()
plt.savefig('../../images/graph_gs.png')
   
```
