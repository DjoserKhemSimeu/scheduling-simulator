```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Comparaison of score between hypercube and random intialization of populations

```python
df=pd.read_csv('data_store/AVGBD_data_r.csv', header=None, names=['Bounded Slowdown'])

plt.figure(figsize=(15, 9))
df
plt.plot(df)
```

```python
# Nombre total de lignes dans le DataFrame
total_lignes = len(df)

# Nombre de lignes dans chaque sous-DataFrame
taille_section = total_lignes // 10

# Liste pour stocker les 10 sous-DataFrames
sous_dataframes = []
plt.figure(figsize=(15, 9))

# Découper le DataFrame en 10 sous-DataFrames
for i in range(10):
    debut_index = i * taille_section
    fin_index = (i + 1) * taille_section
    if i == 9:
        fin_index = total_lignes
    sous_df = df.iloc[debut_index:fin_index].reset_index(drop=True)  # Réinitialiser l'index
    sous_dataframes.append(sous_df)

# Affichage des 10 sous-DataFrames
for i, sous_df in enumerate(sous_dataframes):
    plt.plot(sous_df, label=f"Tuple {i+1}")

# Ajout de la légende
plt.xscale('linear')  # échelle linéaire pour l'axe des x
plt.yscale('linear')  # échelle linéaire pour l'axe des y
plt.title('Random shuffle evaluation')
plt.xlabel('Generation')
plt.ylabel('Bounded Slowdown')
plt.legend(loc='upper right', fontsize='small')

# Affichage du graphique
plt.show()
plt.savefig('../../images/graph_gs.png')
   
```

## Représentation of the two initializations methods in 2D

```python
init_pop_h=pd.read_csv('data_store/pop_init_h.csv', header=None)

plt.figure(figsize=(15, 9))
init_pop_h
```

```python
init_pop_r=pd.read_csv('data_store/pop_init_r.csv', header=None)

plt.figure(figsize=(15, 9))
init_pop_r
```

```python
from sklearn.decomposition import PCA

n_compo=2
pca_h =PCA(n_components=n_compo)
lower_dim_data_h =pca_h.fit_transform(init_pop_h)
pca_r =PCA(n_components=n_compo)
lower_dim_data_r =pca_r.fit_transform(init_pop_r)
lower_dim_data_h
```

```python
plt.scatter(lower_dim_data_h[:,0],lower_dim_data_h[:,1])
```

```python
plt.scatter(lower_dim_data_r[:,0],lower_dim_data_r[:,1])
```

```python
from random import shuffle,randint,random
from scipy.stats import qmc
import numpy as np
size_obs=40
size_test=32
test1=np.arange(size_test)
shuffle(test1)
res_r_x=list()
res_r_y=list()
for i in range (0,size_obs):
    shuffle(test1)
    res_r_x.append(test1[0])
    res_r_y.append(test1[1])
print (res_r_x)
```

```python
from pyDOE3 import *

lhs=lhs(2,samples=size_obs,criterion = 'center')
print(lhs)

res_h_x=list()
res_h_y=list()
for i in range(0,size_obs):
    prob=lhs[i]
    copy=[]
    for j in range(0,2):
        idx=randint(0,size_test - 1)
        p=random()
        
        while (np.isin(idx,copy) or p>prob[j]) :
            idx=randint(0, size_test - 1)
            p=random()
        copy.append(idx)
    res_h_x.append(copy[0])
    res_h_y.append(copy[1])
    
print (res_h_x)
    
```

```python
plt.scatter(res_r_x,res_r_y)
```

```python
plt.scatter(res_h_x,res_h_y)
```

## Observation of the reproductibility of the results with different seed

```python
data_seed_h=pd.read_csv('data_store/AVGBD_data_h_seed.csv', header=None, names=['Bounded Slowdown'])
plt.plot(data_seed_h)
data_seed_h.shape
```

```python
seed_list=[42,23,32,15,234,898,747,45,14,1]
```

```python
# Nombre total de lignes dans le DataFrame
total_lignes = len(data_seed_r)

# Nombre de lignes dans chaque sous-DataFrame
taille_section = total_lignes // 10

# Liste pour stocker les 10 sous-DataFrames
sous_dataframes = []
plt.figure(figsize=(15, 9))

# Découper le DataFrame en 10 sous-DataFrames
for i in range(10):
    debut_index = i * taille_section
    fin_index = (i + 1) * taille_section
    if i == 9:
        fin_index = total_lignes
    sous_df = data_seed_r.iloc[debut_index:fin_index].reset_index(drop=True)  # Réinitialiser l'index
    sous_dataframes.append(sous_df)

# Affichage des 10 sous-DataFrames
for i, sous_df in enumerate(sous_dataframes):
    plt.plot(sous_df[:150], label=f"Seed {seed_list[i]}")

# Ajout de la légende
plt.xscale('linear')  # échelle linéaire pour l'axe des x
plt.yscale('linear')  # échelle linéaire pour l'axe des y
plt.title('Random shuffle with multiple seed')
plt.xlabel('Generation')
plt.ylabel('Bounded Slowdown')
plt.legend(loc='upper right', fontsize='small')

# Affichage du graphique
plt.show()
```
