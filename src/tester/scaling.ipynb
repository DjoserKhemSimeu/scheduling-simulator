```python
import re

class ReaderSWF:
    def __init__(self, filename):
        self.filename = filename
        self.number_of_jobs = None
        self.number_of_processors = None
        self.jobs_info = {'p': [], '~p': [], 'q': [], 'r': []}
        self.read_and_extract_data()

    def read_and_extract_data(self):
        with open(self.filename, 'r') as reader:
            jobs_counter = 0
            for line in reader.readlines():
                
                row = re.split(" +", line.lstrip(" "))
                row[-1] = row[-1].rstrip()


                if row[0].startswith(";"):
                    if "MaxProcs:" in row:
                        self.number_of_processors = int(row[row.index("MaxProcs:") + 1])
                    elif "MaxNodes:" in row:
                        self.number_of_processors = int(row[row.index("MaxNodes:") + 1])
                    continue

                # row[4] is q, row[3] is p, row[8] is ~p, and row[1] is r
                if float(row[4]) > 0 and float(row[4]) <= self.number_of_processors and float(row[3]) > 0 and float(row[8]) > 0:
                    jobs_counter += 1
                    self.jobs_info['p'].append(float(row[3]))
                    self.jobs_info['~p'].append(float(row[8]))
                    self.jobs_info['q'].append(float(row[4]))
                    self.jobs_info['r'].append(float(row[1]))
            
            self.number_of_jobs = jobs_counter
    def write_data(self, filename):
        with open(self.filename, 'r') as reader:
            with open(filename, 'w') as writer:
                jobs_counter = 0
                for line in reader.readlines():
                    
                    row = re.split(" +", line.lstrip(" "))
                    row[-1] = row[-1].rstrip()
                    if not(row[0].startswith(";")) and float(row[4]) > 0 and float(row[4]) <= self.number_of_processors and float(row[3]) > 0 and float(row[8]) > 0:
                        
                        row[3]=str(self.jobs_info['p'][jobs_counter])
                        row[8]=str(self.jobs_info['~p'][jobs_counter])
                        row[4]=str(self.jobs_info['q'][jobs_counter])
                        row[1]=str(self.jobs_info['r'][jobs_counter])
                        jobs_counter += 1
                    new_row=[]
                    for x in row :
                        new_row.append(x)
                        new_row.append(" ")
                    new_row[len(new_row)-1]='\n'
                    writer.writelines(new_row)

                
                
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler

df1=ReaderSWF("../../data/workloads/lublin_256.swf")
df2=ReaderSWF("../../data/workloads/lublin_256.swf")
df=pd.DataFrame.from_dict(df1.jobs_info)
scaler1=StandardScaler()
scaler2=MinMaxScaler()
arr2=scaler1.fit_transform(df)
arr3=scaler2.fit_transform(df)

df_norm_std=pd.DataFrame(arr2, columns = ["p","~p","q","r"])
df_norm_minmax=pd.DataFrame(arr3, columns = ["p","~p","q","r"])
dict_minmax=df_norm_minmax.to_dict('list')
dict_std=df_norm_std.to_dict('list')
df1.jobs_info=dict_std
df2.jobs_info=dict_minmax
df1.write_data("../../data/workloads/lublin_256_norm_std.swf")
df2.write_data("../../data/workloads/lublin_256_norm_minmax.swf")
df3=ReaderSWF("../../data/workloads/lublin_256_norm_std.swf")
dict=pd.DataFrame.from_dict(df3.jobs_info)
dict
```

```python
import pandas as pd
import numpy as np
import math
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("../../data/global_training_data_GA.csv")
df.columns=["p","q","r","y"]
y_save=df["y"]

df=df[["p","q","r"]]
# Calcul de la moyenne du nombre de tÃ¢ches (arrondie Ã  l'entier supÃ©rieur)
num_tasks = [2835, 3836, 4089, 2987, 3079, 3446, 3152, 3554, 4560, 3824,
             2950, 4127, 3257, 2718, 3157, 4559, 3551, 3559, 3062, 2726,
             3376, 2773]
mean_tasks = math.ceil(np.mean(num_tasks))

# Fonction pour diviser un DataFrame en morceaux de taille Ã©gale
def split_dataframe(df, chunk_size):
    chunks = [df[i:i + chunk_size] for i in range(0, len(df), chunk_size)]
    return chunks

# Diviser le DataFrame en morceaux de taille Ã©gale Ã  la moyenne du nombre de tÃ¢ches
chunks = split_dataframe(df, mean_tasks)

# Appliquer la normalisation standard sur chaque DataFrame
scaler = StandardScaler()
normalized_chunks = [pd.DataFrame(scaler.fit_transform(chunk), columns=chunk.columns) for chunk in chunks]

# ConcatÃ©ner de nouveau les morceaux normalisÃ©s
normalized_df = pd.concat(normalized_chunks, ignore_index=True)
normalized_df=pd.concat([normalized_df,y_save],axis=1)
# Affichage du DataFrame normalisÃ©
normalized_df.to_csv("../../data/global_training_data_GA_norm_std.csv",index=False)
normalized_df
```
